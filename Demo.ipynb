### VADER Sentiment Over Time

```python
from plot_utilities import plot_avgs
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
vader_df = pd.read_csv("sentiment-tagging/data/vader_compound_dailies.csv", parse_dates=['day'], infer_datetime_format=True)
plot_avgs(vader_df, plt, 'average_compound', "Sentiment", False, True, True)

plt.title('Sentiment of Tweets about COVID-19 over Time in the US', fontsize=17)
plt.xlabel('Date', fontsize=14)
plt.ylabel('Average Sentiment', fontsize=14)
plt.legend()

plt.show()
```
